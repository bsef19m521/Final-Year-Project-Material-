## WindDir9am and RainTomorrow

```python
ct_counts = df.groupby(['WindDir9am', 'RainTomorrow']).size()
ct_counts = ct_counts.reset_index(name = 'count')
ct_counts = ct_counts.pivot(index = 'RainTomorrow', columns = 'WindDir9am', values = 'count')
```

```python
sns.heatmap(ct_counts)
```

```python
pd.to_datetime(df.Date).dt.year
```

```python
plt.title("Number of rows per Year")
sns.countplot(y=pd.to_datetime(df.Date).dt.year)
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
df['Year'] = df['Date'].dt.year
```

#### Date in Year and RainTomorrow

```python
ct_counts1 = df.groupby(['Year', 'RainTomorrow']).size()
ct_counts1 = ct_counts1.reset_index(name = 'count')
ct_counts1 = ct_counts1.pivot(index = 'RainTomorrow', columns = 'Year', values = 'count')
```

```python
sns.heatmap(ct_counts1)
```
